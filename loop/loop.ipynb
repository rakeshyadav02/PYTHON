## Write a program to accept the numbers continuously; when we enter"-1â€ it should stop accepting the input and at the same time display the number of even numbers and odd numbers that were entered.
Sample Input
Enter the number = 56
Enter the number = 5
Enter the number = 26
Enter the number = 45
Enter the number = 13
Enter the number = 11
Enter the number = 16
Enter the number = 19
Enter the number = 48
Enter the number = -1
Sample Output
Number of Even Numbers Entered = 4
Number of Odd Numbers Entered = 5


```python
even_count = 0
odd_count = 0

while True:
    number = int(input("Enter the number: "))

    if number == -1:
        break

    if number % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Number of Even Numbers Entered:", even_count)
print("Number of Odd Numbers Entered:", odd_count)
```

## Write a program to display the factors of a given number.
Sample Input:56,
sample output:
1, 2, 4, 7, 8, 14, 28, 56



```python

number = int(input("Enter a number: "))


for i in range(1, number+1):
    if number % i == 0:
        print(i,end=",")
print('\b')
```

## Write a program to display the prime in the given range. 
Sample Input
1
10
Sample Output
2 , 3 , 5 , 7


```python
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))


for number in range(start, end + 1):
    if number > 1:
        for i in range(2, number):
            if (number % i) == 0:
                break
        else:
            print(number,end=",")
print('\b')
```
